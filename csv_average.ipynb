```python
import glob
import pandas as pd
import numpy as np
import csv
```

```python
path = "D:/ml/kaggle/State Farm Distracted Driver Detection/result/*"
```

```python
files = glob.glob(path) # ディレクトリにあるファイル一覧を取得
```

```python
file_count = 0
data = {}
for file in files:
    data[file_count] = pd.read_csv(file, engine='python')
    file_count+=1
```

```python
data[0].head()
```

```python
data[1].head()
```

```python
col_c = {}
submit_style = []
for col_num in range(10):
    col_c[col_num] = data[0]["c"+str(col_num)]
    for i in range(1, file_count):
        col_c[col_num] = np.c_[col_c[col_num], data[i]["c"+str(col_num)]]
    if col_num == 0:
        submit_style = np.mean(col_c[col_num], axis=1)
    else:
        submit_style = np.c_[submit_style, np.mean(col_c[col_num], axis=1)]
```

```python
submit_style = np.c_[submit_style, data[0]["img"]]
```

```python
df = pd.DataFrame(data=submit_style, columns=["c0","c1","c2","c3","c4","c5","c6","c7","c8","c9","img"])
```

```python
df.to_csv('D:/ml/kaggle/State Farm Distracted Driver Detection/all/average_submit.csv', index=False)
```
